```python
# Install datasets Library (if not already installed)
# !pip install -U datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_date'] = df['job_posted_date'].dt.date

df.head()
```

```python
df['job_posted_month'] = pd.to_datetime(df['job_posted_date']).dt.month
```

```python
df.sort_values(by='job_posted_date', inplace=True)
```
